```python
import numpy as np
import SWAMPE as sw
```

```python
a = 5.82*10**(7) #Saturn radius m

omega = 1.63*10**(-4) #rotation rate rad/s

Phibar = 4*10**(6) #ref geopotential height m^2/s^2


dt = 1200 #timestep (s)
tmax = 24 # number of timesteps for simulation

M=42 #spectral resolution

DPhieq = Phibar
taurad = 5*3600*24 #one day radiative timescale
taudrag = 10*3600*24 #ten day radiative timescale

plotflag=True
plotfreq = 22 #output: every n time steps

sw.run_model(M, dt, tmax, Phibar, omega, a, taurad=taurad, taudrag=taudrag, DPhieq=DPhieq, plotflag=plotflag, plotfreq=plotfreq) #, saveflag=True, savefreq=30, custompath='C:/Users/Furey/skewl/') #, verbose=False, timeunits='Seconds (s)')
```
